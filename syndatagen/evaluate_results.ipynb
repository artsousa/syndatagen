```python
import os
import pandas as pd

PATH = os.path.join('syndatagen', 'results')
NAMES = ['mistral', 'llama8b_dolphin', 'llama8b_v4', 'llama8b']

results = []
for name in NAMES:
    results.append(pd.read_csv(os.path.join(PATH,  f'result_{name}.csv')))
results[0].head(3)
```

```python
import ast

COLUMNS = ['constraint', 'deepeen', 'concretizing', 'reasoning']
for i in range(1):
    for df in results:
        print(f"\nmodel: {df.iloc[i]['model']}; query: '{df.iloc[i]['query']}'")      
        for col in COLUMNS:
            pair = ast.literal_eval(df.iloc[i][col])
            print(f"[{col}]-[{pair['instruction']}]")
            print(f"{pair['positive'].strip()}")
        print('\n\n')
    break
```
